### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Pada dataset ini terdapat 1 kolom outliers, yaitu pada kolom acousticness dengan outliers sebesar 1.0%. Maka akan dilakukan penanganan outliers pada kolom tersebut.

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

```python
penjualan = df.groupby('primary_genre')['total_streams_billions'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(10,6))
sns.barplot(x=penjualan.index, y=penjualan.values, palette='pastel', hue=penjualan.index, legend=False)
plt.title('Perbandingan 5 Genre dengan Total Streams Tertinggi')
plt.xlabel('Genre')
plt.ylabel('Total Streams (Billions)')
plt.xticks(rotation=45)
plt.show()
```

Genre Pop memiliki total streams paling tinggi dibandingkan genre lainnya, menunjukkan bahwa genre ini paling mendominasi popularitas lagu pada dataset Spotify Wrapped 2025. Sementara itu, Indie Pop, Hip-Hop, Synth-Pop, dan Alt Rock memiliki total streams yang lebih rendah dibandingkan Pop.

## **Composition/Komposisi**

```python
variabel = df['artist_country'].value_counts().sort_values(ascending=False)
variabel2 = variabel.head(5)

plt.figure(figsize=(10, 8))
variabel2.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)
plt.title('Komposisi Negara Artis Teratas')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Artis asal United States mendominasi dataset dengan persentase lebih dari 60%, diikuti oleh United Kingdom sekitar 20%. Hal ini menunjukkan bahwa industri musik global pada dataset ini masih didominasi oleh artis dari negara-negara Barat.

## **Distribution/Distribusi**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['total_streams_billions'], bins=20, kde=True, color='#FFD65A')
plt.title('Distribusi Streams 2025')
plt.xlabel('total_streams_billions')
plt.ylabel('Frekuensi')
plt.show()
```

Sebagian besar lagu memiliki total streams pada kisaran 1,5 hingga 3,5 miliar streams. Distribusi data terlihat tidak merata dan sedikit condong ke kanan (right skewed), karena terdapat beberapa lagu dengan streams sangat tinggi dibandingkan lagu lainnya.

## **Relationship/Hubungan**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['bpm', 'energy']].corr(),
            annot=True,
            cmap='Pastel1',
            fmt='.2f')

plt.title('Korelasi antara BPM dan Energy')
plt.show()
```

Nilai korelasi sebesar 0,54 menunjukkan adanya hubungan positif sedang antara BPM dan energy. Hal ini berarti lagu dengan tempo lebih cepat cenderung memiliki tingkat energi yang lebih tinggi.
